```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

data = data.dropna()

# Standardizing the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
data_scaled
```

```python
data.columns
```

```python
# Choosing the number of clusters (k)
# Replace with your chosen value of k
k = 3

# K-Means clustering
kmeans = KMeans(n_clusters=k, random_state=0)
clusters = kmeans.fit_predict(data)
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Assigning different colors to each cluster
colors = ['blue', 'green', 'red', 'purple', 'orange', 'yellow']  # Extend if more clusters
for i in range(k):
    ax.scatter(data_scaled[clusters == i, 0], data_scaled[clusters == i, 1], data_scaled[clusters == i, 2], 
               s=50, c=colors[i], label=f'Cluster {i+1}')

ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
plt.title('3D Visualization of Clusters')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming data is pre-scaled and has columns 'feature1', 'feature2', 'feature3'
# Replace these with your actual column names if different

# K-Means clustering
k = 3  # Adjust the number of clusters as needed
kmeans = KMeans(n_clusters=k, random_state=0)
clusters = kmeans.fit_predict(data)

import numpy as np

# Convert DataFrame to NumPy array for plotting
data_np = data.to_numpy()

# Visualization
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Set labels with the original feature names
ax.set_xlabel(data.columns[0])
ax.set_ylabel(data.columns[1])
ax.set_zlabel(data.columns[2])

# Plotting the clusters
colors = ['blue', 'green', 'red', 'purple', 'orange', 'yellow']  # Extend if more clusters
for i in range(k):
    ax.scatter(data_np[clusters == i, 0], data_np[clusters == i, 1], data_np[clusters == i, 2], 
               s=50, c=colors[i], label=f'Cluster {i+1}')

plt.title('3D Visualization of Clusters')
plt.legend()
plt.show()

```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame

# Generate the linkage matrix
Z = sch.linkage(data, method='ward')  # 'ward' is one method of hierarchical clustering

# Plotting dendrogram
plt.figure(figsize=(12, 7))
dendrogram = sch.dendrogram(Z)
plt.title('Dendrogram')
plt.xlabel('Observations')
plt.ylabel('Distance')
plt.show()

```

```python
# Choose the number of clusters from dendrogram
n_clusters = 3  # Adjust based on your choice

# Applying Agglomerative Clustering
hc = AgglomerativeClustering(n_clusters=n_clusters, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(data)

# Optionally, you can visualize the clusters (in 2D or 3D depending on your data)
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Assuming 'data' is your DataFrame and 'y_hc' contains the cluster labels
# from hierarchical clustering

# Creating a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Set the colors for each cluster
colors = ['blue', 'green', 'red', 'purple', 'orange', 'yellow']  # Extend if more clusters

# Plot each cluster
for i in range(n_clusters):
    ax.scatter(data[y_hc == i].iloc[:, 0], 
               data[y_hc == i].iloc[:, 1], 
               data[y_hc == i].iloc[:, 2], 
               s=50, c=colors[i], label=f'Cluster {i+1}')

# Setting labels with original feature names
ax.set_xlabel(data.columns[0])
ax.set_ylabel(data.columns[1])
ax.set_zlabel(data.columns[2])

plt.title('3D Visualization of Hierarchical Clustering')
plt.legend()
plt.show()
```
